```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
mouse_data.head()

clinical_data = pd.read_csv(clinical_trial_data_to_load)
clinical_data.head()

# Combine the data into a single dataset
combined_data = pd.merge(clinical_data,mouse_data,how="left",on=["Mouse ID","Mouse ID"])


# Display the data table for preview
combined_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_mean=combined_data.groupby(["Drug","Timepoint"]).mean()["Tumor Volume (mm3)"]

# Convert to DataFrame
tumor_mean=pd.DataFrame(tumor_mean)
tumor_mean=tumor_mean.reset_index()

# Preview DataFrame
tumor_mean.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_error= combined_data.groupby(["Drug","Timepoint"]).sem()["Tumor Volume (mm3)"]

# Convert to DataFrame
tumor_error=pd.DataFrame(tumor_error)
tumor_error=tumor_error.reset_index()

# Preview DataFrame
tumor_error.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

tumor_mean=tumor_mean.reset_index()
tumor_pivot_mean=tumor_mean.pivot(index="Timepoint",columns="Drug")["Tumor Volume (mm3)"]

tumor_error=tumor_error.reset_index()

tumor_pivot_sem=tumor_mean.pivot(index="Timepoint", columns="Drug")["Tumor Volume (mm3)"]

# Preview that Reformatting worked
tumor_pivot_mean.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(tumor_pivot_mean.index, tumor_pivot_mean["Capomulin"], yerr=tumor_pivot_sem["Capomulin"],
             color="r", marker="o", markersize=5, linestyle="dashed", linewidth=.50, label="Capomulin")

plt.errorbar(tumor_pivot_mean.index, tumor_pivot_mean["Infubinol"], yerr=tumor_pivot_sem["Infubinol"],
             color="b", marker="^", markersize=5, linestyle="dashed", linewidth=.50, label="Infubinol" )

plt.errorbar(tumor_pivot_mean.index, tumor_pivot_mean["Ketapril"], yerr=tumor_pivot_sem["Ketapril"],
             color="g", marker="s", markersize=5, linestyle="dashed", linewidth=.50, label="Ketapril")

plt.errorbar(tumor_pivot_mean.index, tumor_pivot_mean["Placebo"], yerr=tumor_pivot_sem["Placebo"],
             color="black", marker="D", markersize=5, linestyle="dashed", linewidth=.50, label="Placebo")

#plt.xlim(0,40)
#plt.ylim(35,75)

plt.legend(loc="best")

plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")


# Save the Figure
plt.savefig("../Images/treatment_response.png")
#plt.show()
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_error= combined_data.groupby(["Drug","Timepoint"]).sem()["Metastatic Sites"]

# Convert to DataFrame
met_error=pd.DataFrame(met_error)
met_error=met_error.reset_index()

# Preview DataFrame
met_error.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
met_mean=met_mean.reset_index()
met_pivot_mean=met_mean.pivot(index="Timepoint",columns="Drug")["Metastatic Sites"]

met_error=met_error.reset_index()
met_pivot_sem=met_mean.pivot(index="Timepoint", columns="Drug")["Metastatic Sites"]


# Preview that Reformatting worked
met_mean.head()
```

```python
# Generate the Plot (with Error Bars)

#tumor_pivot_mean.index, tumor_pivot_mean["Placebo"], yerr=tumor_pivot_sem["Placebo"],

plt.errorbar(met_pivot_sem.index, met_pivot_mean["Capomulin"], yerr=met_pivot_sem["Capomulin"],
             color="r", marker="o", markersize=5, linestyle="dashed", linewidth=.50, label="Capomulin")

plt.errorbar(met_pivot_sem.index, met_pivot_mean["Infubinol"], yerr=met_pivot_sem["Infubinol"],
             color="b", marker="^", markersize=5, linestyle="dashed", linewidth=.50, label="Infubinol" )

plt.errorbar(met_pivot_sem.index, met_pivot_mean["Ketapril"], yerr=met_pivot_sem["Ketapril"],
             color="g", marker="s", markersize=5, linestyle="dashed", linewidth=.50, label="Ketapril")

plt.errorbar(met_pivot_sem.index, met_pivot_mean["Placebo"], yerr=met_pivot_sem["Placebo"],
             color="black", marker="D", markersize=5, linestyle="dashed", linewidth=.50, label="Placebo")

#plt.xlim(0,40)
#plt.ylim(0.0,4.0)

plt.legend(loc="best")

plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Metastatic Sites")

# Save the Figure

plt.savefig('../Images/treatment_response.png')
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric
mice_count=combined_data.groupby(["Drug","Timepoint"]).count()["Mouse ID"]
# Convert to DataFrame
mice_count=pd.DataFrame(mice_count)
mice_count=mice_count.reset_index()
# Preview DataFrame
mice_count.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
#mice_count=mice_count.reset_index()
mice_count=mice_count.pivot(index="Timepoint",columns="Drug")["Mouse ID"]
```

```python
mouse_total = mice_count.max()
mouse_percent = mice_count.copy()

for j in range(len(mouse_percent.columns)):
    for i in range(len(mouse_percent)):
        mouse_percent.iloc[i,j] = (mouse_percent.iloc[i,j]/mouse_total[j])*100
        

# Preview the Data Frame
mouse_percent.head()
```

```python
# Generate the Plot (Accounting for percentages)
plt.errorbar(mouse_percent.index, mouse_percent["Capomulin"],color="r", 
             marker="o", markersize=5, linestyle="dashed", linewidth=.50, label="Capomulin")

plt.errorbar(mouse_percent.index,mouse_percent["Infubinol"],color="b",
             marker="^", markersize=5, linestyle="dashed", linewidth=.50, label="Infubinol" )

plt.errorbar(mouse_percent.index, mouse_percent["Ketapril"], 
             color="g", marker="s", markersize=5, linestyle="dashed", linewidth=.50, label="Ketapril")

plt.errorbar(mouse_percent.index, mouse_percent["Placebo"],
             color="black", marker="D", markersize=5, linestyle="dashed", linewidth=.50, label="Placebo")

plt.legend(loc="best")

plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")

# Save the Figure
plt.savefig('../Images/survival_rates.png')
```

```python
# Show the Figure
plt.show()
```
